```python
import os
import json
import pandas as pd

data_dir = "data"

rows = []

for folder_name in os.listdir(data_dir):
    folder_path = os.path.join(data_dir, folder_name)

    if os.path.isdir(folder_path):
        claim_number = folder_name.replace("claim ", "").strip()
        answer_path = os.path.join(folder_path, "answer.json")

        if os.path.exists(answer_path):
            with open(answer_path, "r") as f:
                data = json.load(f)

            decision = data.get("decision", "")
            explanation = data.get("explanation", "")
            acceptable_decision = data.get("acceptable_decision", "")

            rows.append({
                "claim_number": claim_number,
                "decision": decision,
                "explanation": explanation,
                "acceptable_decision": acceptable_decision
            })

df = pd.DataFrame(rows)

df.to_csv("claim_summary.csv", index=False)
```
